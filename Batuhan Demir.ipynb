```python
import os
Pizza_tabanları = ["Klasik", "Margarita", "TürkPizza", "Sade"]
Pizza_sosları = ["Zeytin", "Mantar", "Keçi Peyniri", "Et", "Soğan", "Mısır"]
Taban_fiyatları = {"Klasik" : 20,
                   "Margarita" : 24, 
                   "TürkPizza":  26, 
                   "Sade" : 18
                   }
Sos_fiyatları = {"Zeytin" : 2, 
                 "Mantar" : 3, 
                 "Keçi Peyniri" : 5, 
                 "Et" : 8, 
                 "Soğan" : 2, 
                 "Mısır" : 1
                 }


def Menu():
    os.system("cls")
    print("Pizza Tabanları: ")
    print(Pizza_tabanları,", ")
    print("Pizza Sosları: ")
    print(Pizza_sosları,", ")
    print("\n\n")


def Siparis_al():
    os.system("cls")
    print("Lütfen malzemelerinizi seçiniz")
    siparis = True
    while siparis:
        os.system("cls")
        Menu()
        Taban = input("Lütfen bir pizza tabanı seçiniz:")
        if Taban not in Pizza_tabanları:
            print(f"Üzgünüm ama {Taban} tabanını restoranımızda kullanılmamakta\n.")
            os.system("pause")
            continue
        Sos = input("Lütfen bir sos seçin: ")
        if Sos not in Pizza_sosları:
            print(f"Üzgünüm  ama {Sos} restoranımızda kullanılmamakta \n.")
            os.system("pause")
            continue

        print(f"Siparişiniz: {Taban} tabanlı ve {Sos} soslu pizza: ")
        siparis = False

    return Taban,Sos

class Siparis:
    def __init__(self):
        Taban,Sos = Siparis_al()
        self.type = Taban
        self.Sos = Sos
        self.TabanFiyatları = Taban_fiyatları[Taban]
        self.SosFiyatları = Sos_fiyatları[Sos]
        self.Tutar = self.TabanFiyatları + self.SosFiyatları


Siparis_ilk = True
Siparisler = []
Siparis_ilk = input("Restoranımıza Hoşgeldiniz, siparişinizi alabilir miyim? (y/n): ")
if Siparis_ilk == "y":
 while Siparis_ilk:
        ilksiparis = Siparis()
        Siparisler.append(ilksiparis)
        yenisiparis = input(" Yeniden siparis vermek ister misiniz? (y/n): ")
        if (yenisiparis) == "n":
            break

 Tutar = 0
 for Siparis in Siparisler:
    Tutar+=Siparis.Tutar

 print("Tutarınız: ",Tutar, "₺\n")

 İsim = input("Lütfen adınız girin: ")
 Soyisim = input("Lütfen soyisminizi girin: ")
 Kimlik_No = input("Lütfen T.C. kimlik numaranızı girin: ")
 Kart_Şifre = input("Lütfen kredi kartı Şifrenizi girin: ")
 print("Siparişiniz için teşekkürler!")
else:
    
   print("iyi günler!")
```
